```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the csv data file
file = "/content/sample_data/clean_data_file.csv"
stu = pd.read_csv(file)
```

```python
#Printing info of the dataset
print(stu.info())
```

```python
#Generating a DataFrame
Stu_data = pd.DataFrame(stu)
```

```python
#Understanding the data and dimensions
print(Stu_data.columns)
print(Stu_data.shape)
print(Stu_data.head())
print(Stu_data.describe())
```

```python
#Visualizing the data using different graph plotting techniques

# 1. Histogram Plot
plt.figure(figsize=(10, 8))
plt.hist(Stu_data['writing score'], color='red')
plt.title('Distribution of Writing Score')
plt.xlabel('Writing Score')
plt.ylabel('Frequency')
```

```python
# 2. Count Plot
# Plotting the frequency of each category in the "parental level of education" feature
plt.figure(figsize=(10, 6))
sns.countplot(y='parental level of education', data=Stu_data, palette='husl')
plt.title('parental level of education')
plt.xlabel('Frequency')
plt.ylabel('parental level of education')
plt.show()
```

```python
# 3. Boxplot
# Showing Parental level of education based on race/ethnicity
plt.figure(figsize=(10, 6))
sns.boxplot(x='race/ethnicity', y='parental level of education', data=Stu_data, palette='bright')
plt.title('Parental level of education based on race/ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Parental level of education')
plt.show()
```

```python
# 4. Pie Plot
# Showing of Race/Ethnicity of Students
plt.figure(figsize=(8, 8))
labels = Stu_data['race/ethnicity'].value_counts().index
sizes = Stu_data['race/ethnicity'].value_counts().values
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=100, colors=['#133955','#ECEBBD'])
plt.title('Race/Ethnicity of Students')
plt.show()
```

```python
# 5.Scatter plot
# Comparing Reading Score and Writing Score
plt.figure(figsize=(8, 6))
sns.scatterplot(x='reading score', y='writing score', hue='test preparation course', data=Stu_data)
plt.title(' Reading Score vs  Writing Score')
plt.xlabel('Reading Score')
plt.ylabel('Writing Score')
plt.show()
```
